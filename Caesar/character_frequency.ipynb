```python
import string

# count letters
frequency_dictionary= {}
for c in string.ascii_uppercase:
  frequency_dictionary[c]= { "count": 0 }

books= ["hamlet.txt", "caesar.txt", "war_n_peace.txt", "tempest.txt"]
for book in books:
  text= open(book, "r").read().upper()
  for c in text:
    if c >= 'A' and c <= 'Z':
      frequency_dictionary[c]["count"]+= 1
```

```python
import matplotlib.pyplot

# bar graph of letters
X= frequency_dictionary.keys()
Y= []
for x in X:
  Y.append(frequency_dictionary[x]["count"])
matplotlib.pyplot.figure(figsize= (12, 8))
matplotlib.pyplot.bar(X, Y);
```

```python
# populate probability (p) section
total_letters= 0
for letter in frequency_dictionary:
  total_letters= total_letters + frequency_dictionary[letter]["count"]

for letter in frequency_dictionary:
  p= frequency_dictionary[letter]["count"] / total_letters
  frequency_dictionary[letter]["p"]= p
```

```python
# write to csv
import csv
with open("frequency_distribution.csv", "w") as f:
  writer= csv.writer(f)
  writer.writerow(["letter", "count"])
  for letter in frequency_dictionary: # O(klgk), k= # letters
    writer.writerow([letter,
                     frequency_dictionary[letter]["count"],
                     frequency_dictionary[letter]["p"]])

import pandas
df = pandas.read_csv("frequency_distribution.csv", sep=",")
df
```
